1 ) Let a as the input as the list of n elements. Find the lcm(least common multiple) of n elements.

Let a = 2 3 5 10

ANSWER:

LCM = 30

````python
a = list(map(int, input().split()))
l = max(a)
v = 1
while True:
  lcm = l*v
  v = v+1
  check = all([lcm%i==0 for i in a])
  if check:
    break
lcm    
````

2) Let n be a positive integer greater than 2. List all prime numbers less than n.

19. Write a code to get 2 integers as input and find the HCF of the 2 integer without using recursion or Euclidean algorithm.

A single line containing 2 integers separated by space.

Output Description:
Print the HCF of the integers.

Sample Input :

2 3

Sample Output :

1

